[문제] 계산기 기능을 가진 클래스 작성
1. 숫자를 객체 속성으로 가지고 있는다(입력 형식 고려)
2. 숫자에 대하여 사칙연산, 평균, 최댓값, 최솟값 등에 관해 알려줄 수 있는 메소드를 가진다


```python
# 클래스 정의
class 숫자만추출():
    def __init__(self): pass
    def _숫자추출(self, 입력숫자): pass

class 계산기(숫자만추출):
    def __init__(self, *입력숫자):
        super().__init__()
        #self.숫자 = [숫자 for 숫자 in 입력숫자]
        self.숫자 = []
        self.숫자아님 = []
        self._숫자_추출(입력숫자)

    def _숫자_가져오기(self, 숫자모음, 반환):
        for 숫자 in 숫자모음:
            if isinstance(숫자, int): 반환.append(숫자)
            else: self.숫자아님.append(숫자)
        return 반환

    def _숫자_추출(self, 입력숫자):
        for 숫자 in 입력숫자:
            if isinstance(숫자, int): self.숫자.append(숫자)
            elif type(숫자) in (list, tuple):
                self.숫자.append(self._숫자_가져오기(숫자, self.숫자))
            else: self.숫자아님.append(숫자)
        if len(self.숫자아님) != 0:
            print(f"입력중에서 숫자가 아닌 것들: {self.숫자아님}")

    def 숫자_재입력(self, *입력숫자):
        self.__init__(*입력숫자)
        #self.숫자 = None
        #self.숫자추출(입력숫자)

    def 숫자_추가(self, *입력숫자):
        self._숫자추출(입력숫자)

    def 더하기(self):
        if len(self.숫자) != 0: return sum(self.숫자); return None

    def 빼기(self):
        if len(self.숫자) != 1:
            반환 = self.숫자[0]
            for 숫자 in self.숫자[1:]:
                반환 = 반환 - 숫자
            return 반환
        return None

    def 곱하기(self):
        if len(self.숫자) != 1:
            반환 = self.숫자[0]
            for 숫자 in self.숫자[1:]:
                반환 = 반환 * 숫자
            return 반환
        return None

    def 나누기(self):
        if len(self.숫자) != 1:
            반환 = self.숫자[0]
            for 숫자 in self.숫자[1:]:
                반환 = 반환 / 숫자
            return 반환
        return None

    def 최댓값(self):
        if len(self.숫자) != 0: return max(self.숫자); return None

    def 최솟값(self):
        if len(self.숫자) != 0: return min(self.숫자); return None
```

```python
계산기1 = 계산기(1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')
계산기1.숫자_재입력(2,1,4,2,-9)
계산기1.숫자_추가(5,7,'b')
print(계산기1.숫자)
print(계산기1.더하기())
print(계산기1.빼기())
print(계산기1.곱하기())
print(계산기1.나누기())
print(계산기1.최댓값())
print(계산기1.최솟값())
```

[문제] 개인정보를 저장하는 전화번호부 클래스 만들기
1. 개인정보: 이름, 전화번호, 주소 등
2. 검색 기능 제공

```python
class 개인정보:
    def __init__(self):
        self.개인정보 = {}

    def _개인정보_입력(self, 개인정보: dict):
        for 정보, 내용 in 개인정보.items():
            self.개인정보[정보] = 내용

    def _개인정보_수정(self, 개인정보: dict):
        for 정보, 내용 in 개인정보.items():
            self.개인정보[정보] = 내용

    def _개인정보_반환(self): return self.개인정보


class 전화번호부(개인정보):
    def __init__(self):
        self.전화번호부 = {}
        #개인정보.__init__(self)
        super().__init__()

    def 사람_추가(self, 개인정보: dict):
        self._개인정보_입력(개인정보)
        이름 = 개인정보["이름"]
        self.전화번호부[이름] = self._개인정보_반환()

    def 전화번호부_반환(self):
        return self.전화번호부

    def 전화번호부_검색(self, 검색어):
        if 검색어 in self.전화번호부.keys():#사람이름으로 검색
            return self.전화번호부[검색어]

        반환 = []
        for 사람, 정보 in self.전화번호부.items():#상세정보로 검색
            # if 검색어 in 정보.values():
            #     반환.append(정보)
            for 세부정보 in 정보.values():
                if 검색어 in 세부정보:
                    반환.append(정보)

        return 반환
```

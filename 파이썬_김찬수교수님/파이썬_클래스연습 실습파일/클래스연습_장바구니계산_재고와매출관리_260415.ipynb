```python
import datetime
import csv
import os

class DB생성관리():
    def __init__(self):
        self._DB = {}
        self._DB_항목 = {"품명":"품명", "단가":"단가", "재고":"재고"}

    def DB_품목추가(self, 품명, 단가, 재고):
        self._DB[품명] = {self._DB_항목["품명"]:품명,
                       self._DB_항목["단가"]:단가,
                       self._DB_항목["재고"]:재고}

    def DB_품목삭제(self, 품명): self._DB.pop(품명)

    def DB_생성_form_csv(self, 파일명):
        with open(파일명, mode='r', encoding='cp949') as f:
            csv_dict_reader = csv.DictReader(f)
            for 정보 in csv_dict_reader:
                self.DB_품목추가(정보["품명"], int(정보["단가"]), int(정보["재고"]))
                #self.DB_추가(정보.get("품명"), 정보.get("단가"), 정보.get("재고"))

    def DB_초기화(self): self._DB = {}

    def DB_반환(self): return self._DB

    def DB_저장_csv(self, 파일명):
        with open(파일명, mode='w', encoding='cp949', newline='') as f:
            csv_writer = csv.writer(f)

            항목 = []
            for i in self._DB_항목.values():
                항목.append(i)
            항목.append("주간판매량")
            csv_writer.writerow(항목)

            for 물건, 정보 in self._DB.items():
                저장내용 = []
                #for 저장항목 in _DB_항목.values():
                for 저장항목 in ["품명", "단가", "재고", "주간판매량"]:
                    #저장내용.append(정보[저장항목])
                    #저장내용.append(정보.get(저장항목))
                    try:
                        저장내용.append(정보[저장항목])
                    except:
                        저장내용.append("None")

                csv_writer.writerow(저장내용)

    def 단가반환(self, 품명): return self._DB[품명][self._DB_항목["단가"]]

    def 재고반환(self, 품명): return self._DB[품명][self._DB_항목["재고"]]

    def 단가수정(self, 품명, 단가): self._DB[품명][self._DB_항목["단가"]] = 단가

    def 재고수정_물품(self, 품명, 변동량): self._DB[품명][self._DB_항목["재고"]] += 변동량

    def 재고수정_물품들(self, 품명들):
        for 물품 in 품명들:
            self.재고수정_물품(물품, -1)

    def 매출기록(self, 품명들):
        파일명 = f"{datetime.date.today()}_매출.csv"
        if not 파일명 in os.listdir():
            with open(파일명, mode='w', encoding='utf-8', newline='') as f:
                csv_writer = csv.writer(f)
                csv_writer.writerow(["품목", "단가"])

        # with open(파일명, mode='a+', encoding='cp949', newline='') as f:
        with open(파일명, mode='a+', encoding='utf-8', newline='') as f:
            csv_writer = csv.writer(f)
            for 물품 in 품명들:
                csv_writer.writerow([물품, self.단가반환(물품)])




```

```python
가게_DB = DB생성관리()
가게_DB.DB_생성_form_csv("재고현황.csv")
가게_DB.DB_저장_csv("재고현황_260416.csv")
```

```python
class 매대계산기(DB생성관리):
    def __init__(self):
        super().__init__()
        self.장바구니_계산액 = 0
        self.장바구니_물건갯수 = 0
        self.장바구니_물품 = []

    def 장바구니계산(self, 장바구니):
        for 물건 in 장바구니:
            self.장바구니_계산액 += self.단가반환(물건)
            self.장바구니_물건갯수 += 1
            self.장바구니_물품.append(물건)

        return self.장바구니_계산액

    def 장바구니_물건빼기(self, 품명):
        self.장바구니_계산액 -= self.단가반환(품명)
        self.장바구니_물건갯수 -= 1
        self.장바구니_물품.remove(품명)
        return self.장바구니_계산액

    def 장바구니_결제(self, 금액):
        self.장바구니_계산액 = 0
        self.장바구니_물건갯수 = 0
        self.재고수정_물품들(self.장바구니_물품)
        self.매출기록(self.장바구니_물품)
```

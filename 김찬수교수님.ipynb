```python
def divisor_2(key, ques):
    if ques == 1:
        print(f"{key}의 약수:1, ", end="")
        i = 2
        while i < key:
            if key % i == 0:
                print(i, end=', ')
            i += 1
        print(key)
    else:
        key_count = 0
        print(f"{key}의 약수 갯수: ", end="")
        i = 1
        while i < key + 1:
            if key % i == 0:
                key_count += 1
            i += 1
        print(key_count)
```

```python
# 컴퓨터와 가위바위보 게임 코딩 5전 3선승제 먼저 이기는 사람 알려주고 축하의 인사 출력
```

```python
import random

def rock_scissor_paper(user_choose, user_win, com_win):
    com_number = random.randint(1,3)
    if com_number == 1:
        com_choose = '가위'
    elif com_number == 2:
        com_choose = '바위'
    else:
        com_choose = '보'
    '''
    com_choose = {1:'가위', 2:'바위', 3:'보'}
    com_choose(com_number)
    '''
    if user_choose == com_choose:
        print(f"비겼습니다! 당신:{user_choose}, 컴퓨터:{com_choose}")
    elif ((user_choose == '가위' and com_choose == '보') or
          (user_choose == '바위' and com_choose == '가위') or
          (user_choose == '보' and com_choose == '바위')):
          print(f"이겼습니다! 당신:{user_choose}, 컴퓨터:{com_choose}")
          user_win += 1
    else:
        print(f"졌습니다! 당신:{user_choose}, 컴퓨터:{com_choose}")
        com_win += 1
    print(f"현재 스코어 - 당신: {user_win}, 컴퓨터: {com_win}\n")
    return user_win, com_win

def main():
    user_win = 0
    com_win = 0

    while user_win < 3 and com_win < 3:
        user_choose = input("가위바위보! (선택지: 가위, 바위, 보)")
        only_choose = ['가위', '바위', '보']

        if user_choose not in only_choose:
            print("그런 건 가위바위보 선택지에 없습니다.")
            continue

        user_win, com_win = rock_scissor_paper(user_choose, user_win, com_win)
    if user_win == 3:
        print("축하합니다! 당신은 컴퓨터를 5판 3선승으로 이겼습니다!")
    elif com_win == 3:
        print("아쉽네요~ 당신은 컴퓨터에게 5판 3선승으로 졌습니다!")

if __name__ == "__main__":
    main()
```

```python
'''
랜덤함수 중 choice() 함수는 임의의 집합에서 무작위로 고를 수 있게끔 해주는 함수이다.
승리 조건을 조건문에 일일이 쓰는게 아니라, 딕셔너리로 승리 조건을 만들면 가독성이 높아진다.
(예: 승리 조건 = {가위:보, 바위:가위, 보:바위})
아니면 그냥 딕셔너리를 두번 써서 다 정의해도됨
{가위:{가위:비김, 바위:짐, 보:이김}, 바위:{가위:이김, 바위:비김, 보:짐}, 보:{가위:짐, 바위:이김, 보:비김}}

'''
# 배열을 통해서도 풀이 가능 --> x y 값에 매칭되는 배열값 가져오기
```

```python
def rsp(user_win, com_win):
    rock_scissor_paper = ['가위', '바위', '보']

    result = {'가위':{'가위':'비김', '바위':'짐', '보':'이김'},
            '바위':{'가위':'이김', '바위':'비김', '보':'짐'},
            '보':{'가위':'짐', '바위':'이김', '보':'비김'}}

    while True:
        user_choose = input("가위바위보! (선택:가위 바위 보)")
        if user_choose in rock_scissor_paper:
            break
        print("올바른 선택지를 입력해 주세요.")

    com_choose = random.choice(rock_scissor_paper)
    rsp_result = result[user_choose][com_choose]
    if rsp_result == '이김':
        user_win += 1
        print(f"{rsp_result}, 나:컴퓨터 = {user_win}:{com_win}")
    elif rsp_result == '짐':
        com_win += 1
        print(f"{rsp_result}, 나:컴퓨터 = {user_win}:{com_win}")
    else:
        print(f"{rsp_result}, 나:컴퓨터 = {user_win}:{com_win}")
    return user_win, com_win
```

```python
import random

def main():
    user_win = 0
    com_win = 0

    while user_win < 3 and com_win < 3:
        user_win, com_win = rsp(user_win, com_win)
    if user_win == 3:
        print("축")
    else:
        print("gg")

if __name__ == "__main__":
    main()
```

```python
import random

key = random.randint(1, 100)

while True:
    user_key = int(input("1부터 100까지의 숫자중 하나를 입력하세요!:"))
    if user_key == key:
        print(f"{user_key}는 정답입니다!!!")
        break
    elif user_key < key:
        print(f"{user_key}는 정답보다 작습니다!")
    else:
        print(f"{user_key}는 정답보다 큽니다!")
```

```python
def divisor(key):
    print(f"{key}의 약수:1, ", end="")
    i = 2
    while i < key:
        if key % i == 0:
            print(i, end=', ')
        i += 1
    print(key)
```

```python
while True:
    key = int(input("정수를 입력하시오:"))
    divisor(key)

    question = input("약수 더 구할래요? y or n:")
    if question == "n":
        break
```

```python
def divisor_2(key, ques):
    if ques == 1:
        print(f"{key}의 약수:1, ", end="")
        i = 2
        while i < key:
            if key % i == 0:
                print(i, end=', ')
            i += 1
        print(key)
    else:
        key_count = 0
        print(f"{key}의 약수 갯수: ", end="")
        i = 1
        while i < key + 1:
            if key % i == 0:
                key_count += 1
            i += 1
        print(key_count)
```

```python
key = int(input("정수를 입력하시오:"))
ques = int(input("입력한 정수의 1. 약수를 볼지, 2. 약수 갯수를 볼지 정하시오:"))

divisor_2(key, ques)
```
